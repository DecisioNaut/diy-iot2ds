# Do-It-Yourself-Parcour: Hobbythek für IoT, Big Data in der Cloud und Data Science

## Der Plan 

- Mit **einfachsten, preiswerten und sogar kostenfreien Mitteln** bauen wir uns 
- eine **komplette integrierte "Teststrecke"**, um 
  - mit *IoT Daten* zu generieren, 
  - diese in die *Cloud* zu bringen, um
  - die Daten dort zu *analysieren* und zu *visualisieren* und um
  - mit einer einfachen *künstlichen Intelligenz* weiteren Nutzen zu generieren.

## Das Experiment

- Wir bauen eine **Wetterstation, deren Daten wir im Internet ansehen können, und detektieren ungewöhnliche Ereignisse in den Daten**

## Das Setup

- Wir bauen eine Wetterstation mit einem **Raspberry Pi** (im Smara-Rack),
- wir schieben die Daten in die Datenablage eines Google Accounts (**Google Drive**),
- wir analysieren und visualisieren die Daten mit dem Tool **Google Data Studio** und
- wir programmieren einen Detektor für Datenausreißer in der Online-Programmierumgebung **Google Colab**

The BME280 Breakout has been design to be used in indoor/outdoor navigation, weather forecasting, home automation, and even personal health and wellness monitoring.
The on-board BME280 sensor measures atmospheric pressure from 30kPa to 110kPa as well as relative humidity and temperature. The breakout provides a 3.3V SPI interface, a 5V tolerant I2C interface (with pull-up resistors to 3.3V), takes measurements at less than 1mA and idles less than 5µA.
